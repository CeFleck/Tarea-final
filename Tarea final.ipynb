# Diferencias entre np.array y np.matrix

Las matrices en numpy son muy utilizadas ya que permiten los trabajos en dimensiones. Existen distintos tipos de creacion dentro de numpy que es con listas anidadas arreglos de numpy o las matrices de numpy.

### np.array

###### Que es un np.array en python?
Este es una coleccion de elementos con una limtante N que se parece mucho a las secuencias de python tales como las listas. Tiene propiedades muy similares con algunas propiedades extras que lo hacen muchisimo mas util. 


```python
arreglo =np.array([1,2,3,4,5,6,7])
print(np.size(arreglo))
print(arreglo)
```

Como podemos ver, su construccion es bastante sencilla y su manipulacion tambien. Todos los elementos tienen que ser homogeneos, o sea el mismo tipo. Si se agregan elementos de diferente tipo, el arreglo sera del tipo que sea mayoria. Esto se trabaja bien a no ser que sea un tipo string.

Por que se utilizan arrays?
Python es un lenguaje dinamico y los loops son muy costosos. Con las matrices se pueden tardar en rangos de milisegundos el visualizar la lista, sin embargo. Numpy cuenta con optimizaciones que hacen que las operaciones sean mucho mas rapidas dejando en el rango de los micro segundos


### np.matrix

###### Que es np.matrix en python?

Esta como los arrays son una coleccion de elementos sin embargo a diferencia de los arreglos, las matrices son una especie multidimencional estrictamente de dos dimensiones. Las matrices por lo genreal utilizan columnas y filas y ambas estan definidas en la clase matrix.

Existen varias formas de operar las matrices. como mencione antes los loops son muy costosos en python sin embargo con numpy se puede hacer sumas o multiplicaciones sencillas.


```python
a=np.matrix([[1,2],[1,2]])
a
```

```python
b=np.matrix([[2,1],[2,1]])
b
```

```python
a+b
```

Como pudimos apreciar es sencillo de implementar.

##Cual es la diferencia entre matrices y arrays?

Las matrices como se menciona anteriormente son bidimensionales estrictamente. no cuentan con la caracteristica de ser tridimensionales o tetradimensionales. 
Los arrays al ser anidados pueden almacenar tantas dimensiones como se desee.

aqui un ejemplo de un array unidimensional y tridimensional


```python
unidim=np.array([1,1,1,1,1,1,1])
```

```python
tridi=np.array([[[1]*4]*4]*4)
tridi
```

```python
matriz=np.matrix([[[1]*4]*4]*4)
```
